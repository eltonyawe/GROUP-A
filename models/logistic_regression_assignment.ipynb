```python
#Creating model that recognises/predicts images which have digits from 0-9,0-5
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn import metrics
%matplotlib inline
# Example: Load from a CSV file
digits = load_digits()
print("Image Data Shape :",digits.data.shape)
print("Label Data Shape :",digits.target.shape)
#Checking whether there are missing values
```

```python
plt.figure(figsize=(12,8))
for index, (image,label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title("Training %i\n"% label,fontsize=20)
```

```python
#Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.23, random_state=2)

#Create an instance of the Logistic Regression model and fit it to the training data.
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#Predicting the digit for the 1st image(Prediction for 1st image)
pred = model.predict(X_test[0].reshape(1,-1))
print("Prediction for 1st image")
pred
```

```python
#Predicting the digits for our 1st ten images
pred = model.predict(X_test[0:10])
print("Prediction for 1st ten images")
pred
```

```python
#Predicting the digits for all of the dataset
pred = model.predict(X_test)
print("Prediction for all images")
pred
```

```python
#model and fit it to the training data.
accuracy = accuracy_score(y_test, pred)
conf_matrix = confusion_matrix(y_test, pred)
class_report = classification_report(y_test, pred)

#Evaluate the model's performance using metrics such as accuracy, confusion matrix, and classification report.
print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)

#Look at the model coefficients to understand the relationship between features and the target.
print('Coefficients:', model.coef_)
```

```python
cm =metrics.confusion_matrix(y_test, pred)
print(cm)
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True, fmt="3f", linewidth=5, square=True, cmap="Blues_r")
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
all_sample_title = 'Accuracy Score :', format(accuracy)
plt.title(all_sample_title,size=15)
```

```python
index=0
misclassifiedIndex = []
for predict, actual in zip(pred,y_test):
    if predict == actual:
        misclassifiedIndex.append(index)
    index += 1    
plt.figure(figsize=(20,3))
for plotIndex, wrong in enumerate(misclassifiedIndex[0:4]):
    plt.subplot(1,4, plotIndex +1)
    plt.imshow(np.reshape(X_test[wrong],(8,8)),cmap=plt.cm.gray )
    plt.title("Predicted : {} , Actual : {} ".format(pred[wrong],y_test[wrong],fontsize=20))
```
